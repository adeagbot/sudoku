```python
def cross(A, B,T):
    "Cross product of elements in A and elements in B."
    return [a+b+T for a in A for b in B]
```

```python
digits   = '123456789'
rows     = 'ABCDEFGHI'
cols     = digits
```

```python
squares  = cross(rows, cols,'')
```

```python
print (squares)
```

```python
unitlist = ([cross(rows, c,'C') for c in cols] +
            [cross(r, cols,'R') for r in rows] +
            [cross(rs, cs,'B') for rs in ('ABC','DEF','GHI') for cs in ('123','456','789')])
```

```python
print (unitlist)
```

```python
units = dict((s, [u for u in unitlist for p in u if s in p[0:2] ]) 
             for s in squares)
```

```python
print (units)
```

```python
print (units['A1'])
```

```python
peers = dict((s, set(sum(units[s],[]))-set([s]))
             for s in squares)
```

```python
print (peers['A1'])
```

```python
print (peers.values())
```

# Sudoku dictionary 

```python
for keys,values in peers.items():
    for v in values:
        if keys not in v[0:2] :
            print (keys+':'+v[0:2]+':'+v[2:] )
```
